```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, binom, poisson, geom
```

```python
# 1. Bernoulli Distribution (requires a probability, p)
p = 0.7  # Example: Probability of success = 0.7
bernoulli_sample = np.random.binomial(n=1, p=p, size=100)  # n=1 for Bernoulli
print("Bernoulli Distribution (p=0.7):", bernoulli_sample)
plt.hist(bernoulli_sample, bins=2, align='left', rwidth=0.8)  # 2 bins for 0 and 1
plt.title("Bernoulli Distribution")
plt.xlabel("Outcome (0 or 1)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Binomial Distribution (requires n trials, probability p)
n_trials = 10
p_success = 0.5
binomial_sample = np.random.binomial(n=n_trials, p=p_success, size=100)
print("Binomial Distribution (n=10, p=0.5):", binomial_sample)
plt.hist(binomial_sample, bins=np.arange(0, n_trials + 2) - 0.5, align='mid', rwidth=0.8)  #centered bins
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Geometric Distribution (requires probability p)
p_success_geom = 0.2
geometric_sample = np.random.geometric(p=p_success_geom, size=100)
print("Geometric Distribution (p=0.2):", geometric_sample)
plt.hist(geometric_sample, bins=np.arange(1, max(geometric_sample) + 2) - 0.5, align='mid', rwidth=0.8)
plt.title("Geometric Distribution")
plt.xlabel("Number of Trials to First Success")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (replace with your actual 'geometric_sample' data)
geometric_sample = np.random.geometric(p=0.2, size=1000)  # Example data

# Convert the data to a Pandas Series (Seaborn works well with Pandas)
import pandas as pd
geometric_sample_series = pd.Series(geometric_sample, name="Geometric Sample") #added name

# Seaborn histogram
sns.histplot(geometric_sample_series,
             bins=np.arange(1, max(geometric_sample) + 2) - 0.5,
             binwidth=1, #added binwidth
             kde=True)
plt.title("Geometric Distribution")
plt.xlabel("Number of Trials to First Success")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Poisson Distribution (requires lambda - average rate)
lambda_poisson = 5
poisson_sample = np.random.poisson(lam=lambda_poisson, size=100)
print("Poisson Distribution (lambda=5):", poisson_sample)
plt.hist(poisson_sample, bins=np.arange(0, max(poisson_sample) + 2) - 0.5, align='mid', rwidth=0.8)
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Normal Distribution (requires mean and standard deviation)
mean_normal = 0
std_normal = 1
x = np.linspace(mean_normal - 3 * std_normal, mean_normal + 3 * std_normal, 1000)
plt.plot(x, norm.pdf(x, mean_normal, std_normal))
plt.title("Normal Distribution (Mean=0, Std=1)")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.show()
```

```python
# 6. Exponential Distribution (requires scale, which is 1/lambda)
lambda_exp = 2
exponential_sample = np.random.exponential(scale=1/lambda_exp, size=1000) #scale is 1/lambda
plt.hist(exponential_sample, bins=50, density=True)
plt.title("Exponential Distribution (lambda=2)")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.show()
```

```python
plt.plot(x, norm.cdf(x, 0, 1))
plt.title("Normal Distribution CDF")
plt.show()
```
